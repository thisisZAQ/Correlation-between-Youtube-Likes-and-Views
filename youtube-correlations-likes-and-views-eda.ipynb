# Introduction 

This notebook explores the correlation between likes and views of trending youtube videos of 10 countries. The ten countries are grouped by their continents. To measure correlation, this notebook employs the use of [Pearson Correlation method](https://www.sciencedirect.com/topics/computer-science/pearson-correlation). 

<a id="2"></a> <br>
# Sorting countries into Continents

```python
#ASIA
in_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/INvideos.csv'))
jap_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/JPvideos.csv'))
kr_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/KRvideos.csv'))

#EUROPE
gb_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/GBvideos.csv'))
de_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/DEvideos.csv'))
fr_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/FRvideos.csv'))
ru_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/RUvideos.csv'))

#NORTH AMERICA
ca_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/CAvideos.csv'))
mx_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/MXvideos.csv'))
us_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/USvideos.csv'))
```

<a id="3"></a> <br>
# Correlations in Asia

```python
#Corelations between Likes and Views on YouTube in ASIA using Pearson Correlation Method

in_yt_corr = pd.DataFrame(in_yt[['views','likes']])
in_yt_corr
In_PearsonCorr= in_yt_corr.corr(method = 'pearson')
plt.ticklabel_format(style='plain')
plt.scatter(x=in_yt_corr['likes'], y=in_yt_corr['views'],c='orange',alpha=0.5)
plt.title('Indian Youtube Correlation between Likes and Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()


jap_yt_corr = pd.DataFrame(jap_yt[['views','likes']])
jap_yt_corr
Jap_PearsonCorr= jap_yt_corr.corr(method = 'pearson')
plt.ticklabel_format(style='plain')
plt.scatter(x=jap_yt_corr['views'], y=jap_yt_corr['likes'],c='violet',alpha=0.3)
plt.title('Japanese Youtube Correlation between Likes and Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()

kr_yt_corr = pd.DataFrame(kr_yt[['views','likes']])
kr_yt_corr
KR_PearsonCorr= kr_yt_corr.corr(method = 'pearson')
plt.ticklabel_format(style='plain')
plt.scatter(x=kr_yt_corr['views'], y=kr_yt_corr['likes'],c='blue',alpha=0.3)
plt.title('Korean Youtube Correlation between Likes and Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()

#Comparing Correlations In ASIA
asia=pd.DataFrame(pd.merge(pd.merge(In_PearsonCorr,KR_PearsonCorr,on='views'),Jap_PearsonCorr,on='views'))
asia
#dropping redundant columns
del asia['views']
asia.rename(columns={'likes_x':'India','likes_y':'Japan','likes':'Korea'},inplace=True)
asia
#heatmap of asian countries correlation between likes and comments
sns.heatmap(asia, vmax=1, square=True, annot=True )
plt.title('Heatmap of Asia')
plt.show()
```

<a id="4"></a> <br>
# Correlations In Europe

```python
#Corelations between Likes and Views on YouTube in EUROPE using Pearson Correlation Method

gb_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/GBvideos.csv'))
de_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/DEvideos.csv'))
fr_yt = pd.DataFrame(pd.read_csv('/kaggle/input/youtube-new/FRvideos.csv'))

#Great Britain
gb_yt_corr = pd.DataFrame(gb_yt[['views','likes']])
gb_yt_corr
Gb_PearsonCorr= gb_yt_corr.corr(method = 'pearson')
plt.ticklabel_format(style='plain')
plt.scatter(x=gb_yt_corr['views'], y=gb_yt_corr['likes'],c='purple',alpha=0.3)
plt.title('Great Britain Youtube Correlation between Likes and Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()

#Germany
de_yt_corr = pd.DataFrame(de_yt[['views','likes']])
de_yt_corr
De_PearsonCorr= de_yt_corr.corr(method = 'pearson')
plt.ticklabel_format(style='plain')
plt.scatter(x=de_yt_corr['views'], y=de_yt_corr['likes'],c='c',alpha=0.3)
plt.title('Germany Youtube Correlation between Likes and Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()

#France 
fr_yt_corr = pd.DataFrame(fr_yt[['views','likes']])
fr_yt_corr
Fr_PearsonCorr= fr_yt_corr.corr(method = 'pearson')
plt.ticklabel_format(style='plain')
plt.scatter(x=fr_yt_corr['views'], y=fr_yt_corr['likes'],c='mediumvioletred',alpha=0.3)
plt.title('France Youtube Correlation between Likes and Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()

#Russia
ru_yt_corr = pd.DataFrame(ru_yt[['views','likes']])
Ru_PearsonCorr = ru_yt_corr.corr(method = 'pearson')
plt.ticklabel_format(style='plain')
plt.scatter(x=ru_yt_corr['views'], y=ru_yt_corr['likes'],c='blue',alpha=0.3)
plt.title('Russia Youtube Correlation between Likes and Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()

#Comparing Correlations In EUROPE
europe=pd.DataFrame(pd.merge(pd.merge(Gb_PearsonCorr,De_PearsonCorr,on='views'),Fr_PearsonCorr,on='views'))

europe
#dropping redundant columns
del europe['views']
europe.rename(columns={'likes_x':'Great Britain','likes_y':'Germany','likes':'France'},inplace=True)
europe
#heatmap of asian countries correlation between likes and views 
sns.heatmap(europe, vmax=1, square=True, annot=True )
plt.title('Heatmap of Europe')
plt.show()
```

<a id="5"></a> <br>
# Correlations in North America

```python
#Corelations between Likes and Views on YouTube in NORTH AMERICA using Pearson Correlation Method

ca_yt_corr = pd.DataFrame(ca_yt[['views','likes']])
ca_yt_corr
Ca_Pcorr= ca_yt_corr.corr(method = 'pearson')
plt.ticklabel_format(style='plain')
plt.scatter(x=ca_yt_corr['views'], y=ca_yt_corr['likes'],c='red',alpha=0.3)
plt.title('Canada Youtube Correlation between Likes and Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()

mx_yt_corr = pd.DataFrame(mx_yt[['views','likes']])
mx_yt_corr
Mex_Pcorr= mx_yt_corr.corr(method = 'pearson')
plt.ticklabel_format(style='plain')
plt.scatter(x=mx_yt_corr['views'], y=mx_yt_corr['likes'],c='green',alpha=0.3)
plt.title('Mexico Youtube Correlation between Likes and Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()

us_yt_corr = pd.DataFrame(us_yt[['views','likes']])
us_yt_corr
US_Pcorr= us_yt_corr.corr(method = 'pearson')
plt.ticklabel_format(style='plain')
plt.scatter(x=us_yt_corr['views'], y=us_yt_corr['likes'],c='darkblue',alpha=0.3)
plt.title('USA Youtube Correlation between Likes and Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()


#Comparing Correlations In NORTH AMERICA
northamerica=pd.DataFrame(pd.merge(pd.merge(Ca_Pcorr, Mex_Pcorr,on='views'),US_Pcorr,on='views'))
northamerica
#dropping redundant columns
del northamerica['views']
northamerica.rename(columns={'likes_x':'Canada','likes_y':'Mexico','likes':'USA'},inplace=True)
#heatmap of north american countries correlation between likes and views 
sns.heatmap(northamerica, vmax=1, square=True, annot=True )
plt.title('Heatmap of North America')
plt.show()
```

# Correlations in Eurasia

```python
#Russia
ru_yt_corr = pd.DataFrame(ru_yt[['views','likes']])
Russia_Pcorr = ru_yt_corr.corr(method = 'pearson')
plt.ticklabel_format(style='plain')
plt.scatter(x=ru_yt_corr['views'], y=ru_yt_corr['likes'],c='blue',alpha=0.3)
plt.title('Russia Youtube Correlation between Likes and Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()

sns.heatmap(Russia_Pcorr,vmax=1, square=True, annot=True)
plt.show()

del Russia_Pcorr['views']
Russia_Pcorr
Russia_Pcorr.drop(["likes"],inplace= True)
Russia_Pcorr
```

# Conclusion 

The value for correlation in every country is more than or equal to 0.80 , thus it signifies a strong positive correlation between likes and views. This implies that with an increase in the number of views, the number of likes also increase.
